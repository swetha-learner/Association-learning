## Association Learning

```python
dataset = [['Bread','Milk','Beer'],
          ['Bread','Diapers','Eggs'],
          ['Milk','Diapers','Beer','Cola'],
          ['Bread','Milk','Diapers','Beer'],
          ['Bread','Milk','Cola']]
```

```python
from mlxtend.preprocessing import TransactionEncoder 
import pandas as pd

te = TransactionEncoder()
te=te.fit(dataset) #reads all unique values in your list
print(te.columns_)
te_ary = te.fit_transform(dataset) #creates a boolean 2d array from dataset
print(te_ary)
```

```python
df = pd.DataFrame(te_ary,columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

#finds frequency of items or itemsets
frequent_itemsets = apriori(df, use_colnames = True, min_support = 0.5)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns  import association_rules

#Forming asso rrules amd returning all performances metrics of those asso
rules = association_rules(frequent_itemsets,metric ='support',min_threshold=0.6)
rules
```

```python
rules[rules['confidence']>0.8]
```

```python
rules[rules['confidence']>rules['support'].max()]
```

```python
rules[rules['support']>rules['confidence'].max()]
```

```python
cartItem = ['Milk']

pred = rules[rules['antecedents'].apply(lambda x: set(cartItem).issubset(set(x)))]
pred
```

```python
rules[rules['antecedents']==set(cartItem)]
```
